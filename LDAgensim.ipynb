```python
import warnings
import pyLDAvis.gensim
from gensim import corpora, models, similarities
warnings.filterwarnings('ignore')
warnings.filterwarnings("ignore",category=DeprecationWarning)

dictionary = corpora.Dictionary.load('/tmp/Gensim/Watches.dict')
corpus = corpora.MmCorpus('/tmp/Gensim/Watches.mm')

# First LDA model with 10 topics, 10 passes, alpha = auto
lda_model = models.LdaModel.load('/tmp/Gensim/Watches_10.model')
for i in lda_model.show_topics():
 print i
followers_data =  pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(followers_data)

```

```python
new_doc = "New silver watch!"
new_vec = dictionary.doc2bow(new_doc.lower().split())
print(new_vec)
lda_model[new_vec]
a = list(sorted(lda_model[new_vec], key=lambda x: x[1]))
print(a[0])
print(a[-1])


```

```python
#We can see that the least and the most related topic to our document. Lets check out what words are most associated with those two topics.
lda_model.print_topic(a[0][0]) #least related
```

```python
lda_model.print_topic(a[-1][0]) #most related
```
